```python
#Importing the users table
import pandas as pd
df = pd.read_csv('data/users.csv')
df.head()
```

```python
#Importing the books table

df_books = pd.read_csv('data/books.csv')
df_books.head()
```

```python
#Importing the authors table
df_authors = pd.read_csv('data/authors.csv')
df_authors.head()
```

```python
#Importing the favourites table

df_fv = pd.read_csv('data/favourites.csv')
df_fv.head()
```

```python
#converting users table to sql
df.to_sql('users', engine, if_exists = 'replace')


```

```python
#converting books table to sql
df_books.to_sql('books', engine, if_exists = 'replace')
```

```python
#converting authors table to sql
df_authors.to_sql('authors', engine, if_exists = 'replace')
```

```python
#converting favourites table to sql
df_fv.to_sql('favourites', engine, if_exists = 'replace')
```
